# ✈️ Top 10 Airlines by Total Incidents (1985–2014)

This visualization highlights the **10 airlines with the highest number of reported incidents** over a 30-year span, from 1985 to 2014. An incident refers to any event that could affect flight safety, such as technical issues or operational errors. While not all incidents are severe, consistently high numbers may point to gaps in maintenance, training, or safety protocols.

Analyzing this data helps identify which airlines had the most operational challenges, and sets the stage for deeper investigation into why certain carriers experienced more incidents than others.


```python
df["total_incidents"] = df["incidents_85_99"] + df["incidents_00_14"]
df["total_fatal_accidents"] = df["fatal_accidents_85_99"] + df["fatal_accidents_00_14"]
df["total_fatalities"] = df["fatalities_85_99"] + df["fatalities_00_14"]

```

```python
sns.barplot(x="total_incidents", y="airline", data=df.sort_values("total_incidents", ascending=False).head(10), palette="mako")
plt.title("Top 10 Airlines by Incidents")
plt.xlabel("Total Incidents")
plt.ylabel("Airline")
plt.show()

```

## 🧨 Distribution of Total Fatalities (1985–2014)

This histogram shows how fatalities were spread across airlines over a 30-year span.  
Most airlines had **very few or no fatalities**, while only a few had **significantly high numbers**.  
It highlights that while aviation is generally safe, a small number of carriers accounted for the worst outcomes.


```python
sns.histplot(df["total_fatalities"], bins=15, kde=True, color="crimson")
plt.title("Distribution of Total Fatalities (1985–2014)")
plt.xlabel("Total Fatalities")
plt.ylabel("Number of Airlines")
plt.show()
```

## ⚠️ Top 10 Most Dangerous Airlines (Custom Score v5)

This visualization ranks airlines using a custom danger score:
> `(Fatalities × 5 + Fatal Accidents × 2 + Incidents) ÷ Airline Size`

By giving **fatalities the highest weight**, this score highlights airlines that pose the greatest safety risk — not just those with more flights.

Scaling by airline size makes it a **fair comparison**, revealing the most dangerous carriers over 30 years.


```python
# Calculate custom danger score (version 5)
df["danger_score"] = ((df["total_fatalities"] * 5) + (df["total_fatal_accidents"] * 2) + df["total_incidents"]) / df["avail_seat_km_per_week"] * 1e9

sns.barplot(x="danger_score", y="airline", data=df.sort_values("danger_score", ascending=False).head(10), palette="rocket")
plt.title("Top 10 Most Dangerous Airlines (Custom Score v5)")
plt.xlabel("Danger Score (Scaled)")
plt.ylabel("Airline")
plt.show()
```

## 🟢 Top 10 Safest Airlines (1985–2014)

This final plot shows the **10 safest airlines**, based on our custom danger score that combines incidents, fatal accidents, and fatalities (adjusted for airline size).

### ✳️ How to Read This:
- **Lighter green bars** → **lower danger score** → **safest**
- **Darker green bars** → still safe, but relatively higher scores than top ones
- All airlines here had **excellent safety records** from 1985 to 2014

This wraps up our safety analysis by highlighting **aviation leaders in safety performance** ✈️✅


```python
# Get the 10 airlines with the lowest danger scores
safest_df = df.sort_values("danger_score").head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x="danger_score", 
    y="airline", 
    data=safest_df, 
    palette="Greens"
)

plt.title("Top 10 Safest Airlines (1985–2014)", fontsize=14)
plt.xlabel("Danger Score (Lower = Safer)")
plt.ylabel("Airline")
plt.tight_layout()
plt.show()
```
